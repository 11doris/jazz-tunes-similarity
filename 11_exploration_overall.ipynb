```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('02c_tune_sql_import.csv', sep=',')
df.columns
```

```python
fig = px.histogram(df, x="year",
                   nbins=100,
                   title='Publishing Date of Tunes',
                   range_x=[1900, max(df['year'])]
                   )
fig.show()
```

```python
fig = px.histogram(df, x="num_bars",
                   title='Number of Bars')
fig.show()
```

```python
fig = px.histogram(df, x="num_bars",
                   title='Number of Bars')
fig.show()
```

```python
keys = df.loc[:, ['tune_key', 'tune_mode', 'cycle_fifths_order']]

keys = keys.value_counts(['tune_key', 'tune_mode']).reset_index()
keys.columns = ['tune_key', 'tune_mode', 'count']


fig = px.treemap(keys,
                 path=['tune_mode', 'tune_key'],
                 values='count')
fig.update_traces(root_color='black')
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```

```python
keys = df.loc[:, ['tune_key', 'tune_mode', 'cycle_fifths_order']]
keys = keys.value_counts(['tune_key', 'tune_mode', 'cycle_fifths_order']).reset_index()
keys = keys.sort_values(['cycle_fifths_order', 'tune_mode'])
keys.rename(columns = {0: 'count'}, inplace = True)
keys = keys.reset_index(drop=True)

#keys
```

```python
fig = px.bar(keys, x='count', y='tune_key',
             color='tune_mode',
             barmode='group',
             orientation='h',
             )
fig.update_layout(yaxis = dict( tickfont = dict(size=8)),
                  #autosize=False,
                  #width=500,
                  #height=500
                  )
```

```python
comp = df['composer']
top_n = comp.value_counts().reset_index().head(15)
top_n.columns = ['Composer', 'count']
top_n = top_n.sort_values('count', ascending=True)

top_n

fig = px.bar(top_n, x='count', y='Composer',
             barmode='group',
             orientation='h',
             )
fig.show()
```
